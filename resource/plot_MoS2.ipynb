```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate
```

```python
with open('./kmesh.dat', 'r') as f:#
    l_tight = f.readlines()
k = [float(line.split()[0]) for line in l_tight[1:]]
Epbe = [float(line.split()[1]) for line in l_tight[1:]]

fig=plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
# k 8 8 8
plt.plot(k,Epbe,marker='o',label='pw ecut=50 Ryd',color='black')

plt.xticks(fontsize=28,color='#000000');
plt.yticks(fontsize=28,color='#000000');
font = {'family':'Verdana','weight':'normal','size':29.5,'color':'#000000'}
plt.xlabel('Kpoints',font)
plt.ylabel('Etot(eV)',font)
plt.legend(fontsize=20)
plt.title(r'$MoS_2$ PBE vs kpoints',fontsize=28);
plt.grid(True)
#plt.savefig("kmesh.png", dpi=750, bbox_inches = 'tight')
```
